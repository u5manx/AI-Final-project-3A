```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
import joblib

df = pd.read_csv("SAP-4000.csv")

df = df.dropna(subset=["Exam_Score"])

df = df.fillna({
    "HoursStudied/Week": df["HoursStudied/Week"].mean(),
    "Attendance(%)": df["Attendance(%)"].mean(),
    "Gender": df["Gender"].mode()[0],
    "Tutoring": df["Tutoring"].mode()[0],
    "Region": df["Region"].mode()[0],
    "Parent Education": df["Parent Education"].mode()[0],
})

X = df[[
    "Gender","HoursStudied/Week","Tutoring","Region",
    "Attendance(%)","Parent Education"
]]

cat_cols = X.select_dtypes(include=["object"]).columns
X[cat_cols] = X[cat_cols].apply(lambda col: pd.factorize(col)[0])

y = df["Exam_Score"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, shuffle=False
)

best_model = GradientBoostingRegressor(random_state=42)
best_model.fit(X_train, y_train)

joblib.dump(best_model, "best_gb_model.pkl")
print("Gradient Boost Model saved!")
```
